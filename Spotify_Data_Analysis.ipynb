```python
#Exploring 10 least popular songs
```

```python
least = tracks.sort_values('popularity',ascending = True)[0:10]
least[['name','popularity']]
```

```python
#Method 1 - Exploring 10 Most popular songs
```

```python
most = tracks.sort_values('popularity',ascending = False)[0:10]
most[['name','popularity']]
```

```python
#Descriptive statistics of tracks
```

```python
tracks.describe().transpose()
```

```python
#Setting release date as index column
```

```python
tracks.set_index('release_date', inplace = True)
tracks.index = pd.to_datetime(tracks.index, format  = 'mixed')
tracks.head(10)
```

```python
#Extracting artist name from 7th row of dataset
```

```python
tracks[['name']].iloc[7]
```

```python
#Converting duration of songs from milliseconds to seconds
```

```python
tracks['duration'] = tracks['duration_ms'].apply(lambda x: round(x/1000))
tracks.drop('duration_ms', inplace = True, axis = 1)
tracks.duration.head()
```

```python
#Heatmap using Pearson correlation method between two variables
```

```python
corr = tracks.drop(['key','mode','explicit','id','name','artists','id_artists'],axis = 1).corr(method = 'pearson')
plt.figure(figsize = (10,5))
map = sns.heatmap(corr, annot = True, cmap = 'coolwarm')
map.set_title('Pearson Correlation Heatmap')
map.set_xticklabels(map.get_xticklabels(), rotation = 45)
```

```python
#Creating a regression plot between danceability and energy
```

```python
plt.figure(figsize = (8,4))
sns.regplot(data = tracks, x = 'energy', y = 'loudness').set_title('Regression_Plot')
```

```python
#Sampling only 5% of the dataset
```

```python
sample = tracks.sample(int(0.005*len(tracks)))
print(len(sample))
```

```python
#Using the sample data for a regression plot between popularity and acousticness
```

```python
plt.figure(figsize = (8,4))
sns.regplot(data = sample, x = 'popularity', y = 'acousticness').set_title('Sample Regression Plot') 
```

```python
#Creating a scatterplot for the same data
```

```python
plt.figure(figsize = (8,4))
sns.scatterplot(data = sample, x = 'popularity', y = 'acousticness',color='#008000').set_title('Sample Regression Plot') 
```

```python
#Creating new dates column in tracks table
```

```python
tracks['dates'] = tracks.index.get_level_values('release_date')
tracks.dates = pd.to_datetime(tracks.dates)
years = tracks.dates.dt.year
tracks.head()
```

```python
#Creating a distribution plot for no. of songs each year
```

```python
sns.displot(years, discrete  = True, kind = 'hist', color = 'r',height = 3, aspect = 3).set(title = 'No. of Songs each year')
```

```python
#Creating a bar plot for duration of songs each year
```

```python
total_dur = tracks.duration
fig, ax = plt.subplots(figsize = (10,6))
fig = sns.barplot(x = years, y = total_dur, ax = ax, errwidth = False).set(title = 'Duration of Songs each year')
plt.xticks(rotation = 90)
```

```python
#Top 5 genres by popularity
```

```python
plt.figure(figsize = (8,4))
top = artists.sort_values('popularity', ascending  = False)[0:5]
sns.barplot(y = 'genres', x = 'popularity', data = top).set(title = 'Top 5 genres by popularity')
sns.set_style(style='darkgrid')
```
